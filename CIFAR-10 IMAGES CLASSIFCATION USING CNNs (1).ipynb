### STEP 1 : IMPORT LIBRARIES AND CALL THE DATASET FROM KERAS


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
from keras.datasets import cifar10
(X_train, y_train) , (X_test,y_test) = cifar10.load_data()
```

### STEP 2: VISUALIZE THE DATASET

```python
# This is for one image in our dataset, and also show its class number, which are arrranged in datasets

i = 2300
plt.imshow(X_train[i])
print(y_train[i])
```

## Now we create a grid of some images randomly

```python
w_grid = 15
l_grid = 15

fig, axes = plt.subplots(l_grid , w_grid ,figsize=(25,25))    # it is for make a grid of length and width in 25 by 25 size 

axes = axes.ravel()       # it is for flaten the images

n_training = len(X_train)

for i in np.arange(0, l_grid*w_grid):
    
    index = np.random.randint(0 , n_training)  # pick  random images from the n_training which are choosing images from X_train 
    axes[i].imshow(X_train[index])
    axes[i].set_title(y_train[index])     # it is for title the class of images
    axes[i].axis('off')                 # it is for the off the value of length and width of cell on the below images
plt.subplots_adjust(hspace=0.4)      # it is for space between row for clear title..
    
```

## Step 3 :: DATA PREPARATION  

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
num_cat = 10
```

```python
y_train  # its convert into categorical value
```

```python
import keras
y_train = keras.utils.to_categorical(y_train , num_cat)
```

```python
y_train     # its change in category...like one hot encoding in ML
```

```python
y_test = keras.utils.to_categorical(y_test , num_cat)
```

```python
y_test
```

```python
X_train    # all images change in pixel number...but its not good for calculation..so apply normalization for good result
```

```python
X_train = X_train/255     # here we devide by 255 and normailze the all value in 0 to 1..
X_test = X_test/255
```

```python
X_test
```

```python
X_train.shape
```

```python
# Now we want to only the shape of images not its count

Input_shape = X_train.shape[1:]
```

```python
Input_shape       # we got only shape of the train dataset
```

## STEP 4 :: TRAIN THE MODEL

```python
from keras.models import Sequential
from keras.layers import Conv2D,MaxPooling2D ,AveragePooling2D ,Dense ,Flatten,Dropout
from keras.optimizers import Adam
from keras.callbacks import TensorBoard
```

```python
cnn_model = Sequential()
cnn_model.add(Conv2D(filters =32 , kernel_size=(3,3),activation ='relu', input_shape =Input_shape))
cnn_model.add(Conv2D(filters =32 , kernel_size=(3,3), activation ='relu'))
cnn_model.add(MaxPooling2D(2,2))
cnn_model.add(Dropout(0.30))

cnn_model = Sequential()
cnn_model.add(Conv2D(filters =64 , kernel_size=(3,3),activation ='relu', input_shape =Input_shape))
cnn_model.add(Conv2D(filters =64 , kernel_size=(3,3), activation ='relu'))
cnn_model.add(MaxPooling2D(2,2))
cnn_model.add(Dropout(0.20))

cnn_model.add(Flatten())

cnn_model.add(Dense(units =512, activation ='relu'))    # this  is ANN hidden layer conneting Neurons 
cnn_model.add(Dense(units=512, activation ='relu'))

cnn_model.add(Dense(units= 10, activation ='softmax'))     # this ANN layer is for output 
```

## Now compile the CNN model of all layers 

```python
cnn_model.compile(loss ='categorical_crossentropy',optimizer= keras.optimizers.rmsprop(lr =0.001),metrics=['accuracy'])
```

### Now Fit the train data into Model

```python
history = cnn_model.fit(X_train ,y_train ,batch_size=32 , epochs=1 ,shuffle=True)
```

## STEP 5:: EVALUATE THE MODEL WITH TESTING DATA


```python
evaluation = cnn_model.evaluate(X_test,y_test)
print('CNN Accuracy: {}'.format(evaluation[1]))
```

### Predicted class of X_test

```python
predicted_classes = cnn_model.predict_classes(X_test)
predicted_classes                                           # this shows in decimal format
```

### Now compare with  the True label this predicted class


```python
y_test                    # this is in binary  format...but predicted is in class number 0 to 9... so change it in previous class
```

```python
y_test = y_test.argmax(1)       # so change it in decimal format

y_test
```

### So Now we visualize the predicted and true class 

```python
Length = 10
Width =10
fig, axes = plt.subplots( Length , Width , figsize =(25,25))
axes = axes.ravel()

for i in np.arange(0, Length*Width):
    axes[i].imshow(X_test[i])
    axes[i].set_title('Predicted Class = {}\n True Class ={}'.format(predicted_classes[i],y_test[i]))
    axes[i].axis('off')

plt.subplots_adjust(wspace=0.7)

```

### Now make the confusion matrix.. for summarize all the predicted and true class

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test,predicted_classes)
cm

plt.figure(figsize=(13,13))
sns.heatmap(cm , annot= True)
```
